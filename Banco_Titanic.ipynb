```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
Titanic = sns.load_dataset("titanic")
```

```python
Titanic.head(3)
```

```python
Titanic.info()
```

```python
from seaborn.axisgrid import jointplot
#2 Variáveis join plot
```

```python
sns.jointplot(x = 'fare', y ='age', data = Titanic)
```

```python
sns.distplot(Titanic['fare'])
```

```python
#Classe dos passageiiros e idade
#Dados categóricos
#a terceira classe é representada por pessoas mais novas
sns.boxplot(x= 'class', y='age', data = Titanic, palette= 'rainbow')
```

```python
#gráfico de distribuição de age e categoria

sns.swarmplot(x= 'class', y='age', data= Titanic, palette = 'Set2')
```

```python
sns.countplot(x="sex", data = Titanic)
```

```python
#Matriz
corr= Titanic.corr()
```

```python
sns.heatmap(corr)
```

```python
corr= Titanic.corr()
sns.heatmap(corr, cmap= 'coolwarm')
plt.title('Titanic.corr()')
```

```python
#Distribuição dividida em categorias
g = sns.FacetGrid(data=Titanic, col ="sex")
g.map(plt.hist, "age")
```
